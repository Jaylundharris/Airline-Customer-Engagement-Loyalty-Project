```python
import pandas as pd
import numpy as np

customer_info = pd.read_csv("../data/raw/test.csv")

customer_info.info()
customer_info.head()
customer_info.isna().sum()
```

```python
customer_info.columns = (
    customer_info.columns 
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

```python
customer_info = customer_info.drop(columns=["unnamed:_0"])
```

```python
customer_info = customer_info.dropna(subset=["arrival_delay_in_minutes"])
```

```python
customer_info.to_csv("../data/processed/customer_info_cleaned.csv", index=False)
```

```python
customer_info = pd.read_csv("../data/processed/customer_info_cleaned.csv")

customer_info.head()
customer_info.info()
customer_info.isna().sum()
```

```python
import sqlite3 
import pandas as pd

conn = sqlite3.connect("../data/processed/flight_customer.db")
```

```python
customer_info.to_sql(
    "customer_info_cleaned", 
    conn, 
    if_exists="replace" 
)
```

```python
cic = """
SELECT 
    *
FROM 
    customer_info_cleaned
"""

customer_info_cleaned = pd.read_sql(cic, conn)
customer_info_cleaned
```
